```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm
import seaborn as sns
```

```python
Speaker_Verfication = pd.read_csv("Speaker_Verfication_Performance_0921_Externalas1.csv")
# only compare ratio and d-vector of flatten
Speaker_Verfication = Speaker_Verfication[(Speaker_Verfication["D-vector"] == "Ratio")|(Speaker_Verfication["D-vector"] == "flatten_1")]
```

# local data

```python
Speaker_Verfication_Local = Speaker_Verfication[Speaker_Verfication["Data"] == "Local Data"]
Speaker_Verfication_Local = Speaker_Verfication_Local.reset_index()
```

```python
sns.lineplot(data = Speaker_Verfication_Local, x="Recall",y="Precision",hue='D-vector')
plt.box(False)
plt.legend(["D-vector: Projection","Ratio"],title="Speaker Verification Method",loc=3)
plt.title("PR Curve of Performance on Local Data")
plt.ylim([0,1])
plt.show()
```

# Digit data - 5 speakers

```python
Speaker_Verfication_Digit_5speaker = Speaker_Verfication[(Speaker_Verfication["Data"] == "Digit Data")&
                                                         (Speaker_Verfication["Speaker Number"] == 5)]
Speaker_Verfication_Digit_5speaker = Speaker_Verfication_Digit_5speaker.reset_index()
```

```python
sns.lineplot(data = Speaker_Verfication_Digit_5speaker, x="Recall",y="Precision",hue='D-vector')
plt.box(False)
plt.legend(["D-vector: Projection","Ratio"],title="Speaker Verification Method",loc=3)
plt.title("PR Curve of Performance on Digit Data \nGroup Model Trained with 5 Speakers")
plt.ylim([0,1])
plt.show()
```

# Digit data - 30 speakers - 40% Training data

```python
Speaker_Verfication_Digit_40_30speaker = Speaker_Verfication[(Speaker_Verfication["Data"] == "Digit Data")&
                                                             (Speaker_Verfication["Train Size"] == "40%")&
                                                             (Speaker_Verfication["Speaker Number"] == 30)]
Speaker_Verfication_Digit_40_30speaker = Speaker_Verfication_Digit_40_30speaker.reset_index()
```

```python
sns.lineplot(data = Speaker_Verfication_Digit_40_30speaker, x="Recall",y="Precision",hue='D-vector')
plt.box(False)
plt.legend(["D-vector: Projection","Ratio"],title="Speaker Verification Method",loc=6)
plt.title("PR Curve of Performance on Digit Data \nGroup Model Trained with 30 Speakers & More Data")
plt.ylim([0,1])
plt.show()
```

# Digit data - 30 speakers - 40% Training data for 1-5 speakers and 10% for 6-30 speakers

```python
Speaker_Verfication_Digit_10_30speaker = Speaker_Verfication[(Speaker_Verfication["Data"] == "Digit Data")&
                                                             (Speaker_Verfication["Train Size"] == "40%+10%")&
                                                             (Speaker_Verfication["Speaker Number"] == 30)]
Speaker_Verfication_Digit_10_30speaker = Speaker_Verfication_Digit_10_30speaker.reset_index()
```
