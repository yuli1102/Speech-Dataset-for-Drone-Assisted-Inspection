```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm
import seaborn as sns
```

# Unauthorized

```python
Speaker_Verfication = pd.read_csv("Speaker_Verfication_Performance_0925_ROC.csv")
# only compare ratio and d-vector of flatten
# Speaker_Verfication = Speaker_Verfication[(Speaker_Verfication["D-vector"] == "Ratio")|(Speaker_Verfication["D-vector"] == "flatten_1")]
# only compare the unauthorized as positive result.
Speaker_Verfication = Speaker_Verfication[(Speaker_Verfication["Positive"] == "Unauthorized")]

```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
Speaker_Verfication_Digit = Speaker_Verfication[(Speaker_Verfication["Speaker Number"] == 5)]
Speaker_Verfication_Digit = Speaker_Verfication_Digit.reset_index()
sns.lineplot(data = Speaker_Verfication_Digit, x="fpr",y="tpr",hue='Method')
plt.box(False)
plt.legend(["D-vector: \nProjection","Ratio"],title="Method",loc=4)
plt.title("5 Speakers")
plt.ylim([0,1])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.subplot(1,3,2)
Speaker_Verfication_Digit = Speaker_Verfication[(Speaker_Verfication["Train Size"] == "40%+10%")&
                                                         (Speaker_Verfication["Speaker Number"] == 30)]
Speaker_Verfication_Digit = Speaker_Verfication_Digit.reset_index()
sns.lineplot(data = Speaker_Verfication_Digit, x="fpr",y="tpr",hue='Method')
plt.box(False)
plt.legend(["D-vector: \nProjection","Ratio"],title="Method",loc=4)
plt.title("30 Speakers with Less data")
plt.ylim([0,1])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.subplot(1,3,3)
Speaker_Verfication_Digit = Speaker_Verfication[(Speaker_Verfication["Train Size"] == "40%")&
                                                         (Speaker_Verfication["Speaker Number"] == 30)]
Speaker_Verfication_Digit = Speaker_Verfication_Digit.reset_index()
sns.lineplot(data = Speaker_Verfication_Digit, x="fpr",y="tpr",hue='Method')
plt.box(False)
plt.legend(["D-vector: \nProjection","Ratio"],title="Method",loc=10, bbox_to_anchor=(0.4,0.6))
plt.title("30 Speakers with More data")
plt.ylim([0,1])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.suptitle("Receiver Operating Characteristic Curve of Performance on Digit Data")
plt.tight_layout()
plt.show()
```
