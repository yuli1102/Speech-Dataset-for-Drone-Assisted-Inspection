# Group model

```python
Perfomance_group = pd.read_csv('Performance_0608_training_Data_Size.csv')
Perfomance_group = Perfomance_group[(Perfomance_group["Size"]=='6')|(Perfomance_group["Size"]=='10')|(Perfomance_group["Size"]=='20&s2_30')]
Perfomance_group
```

```python
Perfomance_single = pd.read_csv('Performance_0608_single_model.csv')
Perfomance_single 
```

```python
Perfomance = pd.concat([Perfomance_single,Perfomance_group])
Perfomance = Perfomance.reset_index(drop=True)
Perfomance
```

```python
Perfomance.columns
```

```python
Command_performance = Perfomance[['Model', 'Size', 'Time','Command_Acc_p15','Command_Acc_p1', 'Command_Acc_p2',
                                  'Command_Acc_p3', 'Command_Acc_p4', 'Command_Acc_p5', 'Acc_p6', 'Acc_p7', 'Acc_p8']]
Partcp_performance = Perfomance[['Model', 'Size', 'Time','Partcp_Acc_p15', 'Partcp_Acc_p1', 'Partcp_Acc_p2', 
                                  'Partcp_Acc_p3','Partcp_Acc_p4', 'Partcp_Acc_p5']]
```

```python
Command_performance
```

```python
Partcp_performance[['Model','Size','Time','Partcp_Acc_p15','Partcp_Acc_p1', 'Partcp_Acc_p2',
                                  'Partcp_Acc_p3', 'Partcp_Acc_p4', 'Partcp_Acc_p5']].groupby(['Model','Size']).mean()
```

```python
Command_performance.loc[Command_performance['Size']=="20&s2_30",'Size']="Mix(20,30,20,20,20)"
Command_performance[['Model','Size','Time','Command_Acc_p15','Command_Acc_p1', 'Command_Acc_p2',
                                  'Command_Acc_p3', 'Command_Acc_p4', 'Command_Acc_p5', 'Acc_p6', 'Acc_p7', 'Acc_p8']].groupby(['Model','Size']).mean()
```

```python
Data = pd.DataFrame(columns=['Model','Test_on','Data_size','Accuracy'])
for i in range(len(Command_performance)):
    data = pd.DataFrame(columns=['Model','Test_on','Data_size','Accuracy'])
    data['Test_on'] = Command_performance.columns[4:-3]
    data['Accuracy'] = Command_performance.transpose().values[4:-3][:,i]
    data['Data_size'] = Command_performance.transpose().values[1][i]
    data['Model'] = Command_performance.transpose().values[0][i]
    Data = Data.append(data)
Data = Data.reset_index(drop =True)
Data.Accuracy = Data.Accuracy.astype(float)
plt.figure(figsize=(6,4))
sns.lineplot(data=Data, x = "Test_on", y = "Accuracy",hue = 'Model',style= "Data_size")
plt.xticks(range(5),["1","2","3","4","5"],rotation = 'horizontal')
plt.ylabel("Accuracy",fontsize=12)
plt.xlabel("Speaker",fontsize=12)
# plt.ylim([0.8,1.01])
plt.title("Effectiveness of pooled training data",fontsize=14)
plt.legend(ncol=1,loc=7, bbox_to_anchor=(1.4,0.5))
plt.box(False)
plt.show()
```

```python
Data = pd.DataFrame(columns=["Source",'Test_on','Data_size','Accuracy'])
for i in range(len(Command_performance)):
    data = pd.DataFrame(columns=['Source','Test_on','Data_size','Accuracy'])
    data['Test_on'] = Command_performance.columns[4:]
    data['Accuracy'] = Command_performance.transpose().values[4:][:,i]
    data['Data_size'] = Command_performance.transpose().values[1][i]
    data['Source'] = Command_performance.transpose().values[0][i]
    Data = Data.append(data)
Data = Data.reset_index(drop =True)
Data.Accuracy = Data.Accuracy.astype(float)
plt.figure(figsize=(6,4))
sns.lineplot(data=Data, x = "Test_on", y = "Accuracy",hue = 'Source',style= "Data_size")
plt.xticks(range(8),["1","2","3","4","5","6","7","8"],rotation = 'horizontal')
plt.ylabel("Accuracy",fontsize=12)
plt.xlabel("Speaker",fontsize=12)
# plt.ylim([0.8,1.01])
plt.title("Effectiveness of pooled training data",fontsize=14)
plt.legend(ncol=1,loc=7, bbox_to_anchor=(1.4,0.5))
plt.box(False)
plt.plot([4.5,4.5],[0.3,1.05],'-.',c="gray")
plt.text(3.1,0.4,"Authorized \nSubjects")
plt.text(4.7,0.4,"New \nSubjects")
plt.show()
```
