# calculate by build in function

```python
confidence = pd.DataFrame(columns=["Size", 'Time-low','Time-high', 'Keyword-low', 'Keyword-high','Speaker-low','Speaker-high'])
confidence['Size'] = Perfomance['Size'].unique() 
for size in Perfomance['Size'].unique():
    confidence.loc[confidence['Size'] == size,['Time-low','Time-high']] =  st.t.interval(alpha=0.95, df=len(Perfomance.loc[Perfomance['Size'] == size,'Time']), loc=np.mean(Perfomance.loc[Perfomance['Size'] == size,'Time']), scale=st.sem(Perfomance.loc[Perfomance['Size'] == size,'Time'])) 
    confidence.loc[confidence['Size'] == size,['Keyword-low','Keyword-high']] =  st.t.interval(alpha=0.95, df=len(Perfomance.loc[Perfomance['Size'] == size,'Command_Acc_p15']), loc=np.mean(Perfomance.loc[Perfomance['Size'] == size,'Command_Acc_p15']), scale=st.sem(Perfomance.loc[Perfomance['Size'] == size,'Command_Acc_p15'])) 
    confidence.loc[confidence['Size'] == size,['Speaker-low','Speaker-high']] =  st.t.interval(alpha=0.95, df=len(Perfomance.loc[Perfomance['Size'] == size,'Partcp_Acc_p15']), loc=np.mean(Perfomance.loc[Perfomance['Size'] == size,'Partcp_Acc_p15']), scale=st.sem(Perfomance.loc[Perfomance['Size'] == size,'Partcp_Acc_p15'])) 
confidence
```

```python
confidence[['Time-low','Time-high', 'Keyword-low', 'Keyword-high','Speaker-low','Speaker-high']] = confidence[['Time-low','Time-high', 'Keyword-low', 'Keyword-high','Speaker-low','Speaker-high']].astype('float')
```

```python
confidence.round(3)
```

# calculate by hand

```python
confidence = pd.DataFrame(columns=["Size", 'Time_mean','Time+_', 'Keyword_mean', 'Keyword+_','Speaker_mean','Speaker+_'])
confidence['Size'] = Perfomance['Size'].unique() 
```

```python
t_value = 2.262 # df=9, alpha = 0.05, two tails
def CI_95(data):
    n=len(data)
    x_bar = data.mean()
    x_std = data.std()
    CI = t_value*x_std/np.sqrt(n)
    return x_bar, CI
```

```python
for size in Perfomance['Size'].unique():
    confidence.loc[confidence['Size'] == size,['Time_mean','Time+_']] = CI_95(Perfomance.loc[Perfomance['Size'] == size,'Time']) 
    confidence.loc[confidence['Size'] == size,['Keyword_mean', 'Keyword+_']] = CI_95(Perfomance.loc[Perfomance['Size'] == size,'Command_Acc_p15']) 
    confidence.loc[confidence['Size'] == size,['Speaker_mean','Speaker+_']] = CI_95(Perfomance.loc[Perfomance['Size'] == size,'Partcp_Acc_p15'])  
confidence
```

```python
confidence[['Time_mean','Time+_', 'Keyword_mean', 'Keyword+_','Speaker_mean','Speaker+_']] = confidence[['Time_mean','Time+_', 'Keyword_mean', 'Keyword+_','Speaker_mean','Speaker+_']].astype('float')
```

```python
confidence.round(4)
```
