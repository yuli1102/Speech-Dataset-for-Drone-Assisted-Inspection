```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Group model

```python
Perfomance = pd.read_csv('Performance_0608_training_Data_Size.csv')
Perfomance
```

```python
Perfomance.columns
```

```python
Command_performance = Perfomance[['Model', 'Size', 'Time','Command_Acc_p15','Command_Acc_p1', 'Command_Acc_p2',
                                  'Command_Acc_p3', 'Command_Acc_p4', 'Command_Acc_p5', 'Acc_p6', 'Acc_p7', 'Acc_p8']]
Partcp_performance = Perfomance[['Model', 'Size', 'Time','Partcp_Acc_p15', 'Partcp_Acc_p1', 'Partcp_Acc_p2', 
                                  'Partcp_Acc_p3','Partcp_Acc_p4', 'Partcp_Acc_p5']]
```

```python
Command_performance
```

```python
Partcp_performance[['Model','Size','Time','Partcp_Acc_p15','Partcp_Acc_p1', 'Partcp_Acc_p2',
                                  'Partcp_Acc_p3', 'Partcp_Acc_p4', 'Partcp_Acc_p5']].groupby('Size').mean()
```

```python
Command_performance[['Model','Size','Time','Command_Acc_p15','Command_Acc_p1', 'Command_Acc_p2',
                                  'Command_Acc_p3', 'Command_Acc_p4', 'Command_Acc_p5', 'Acc_p6', 'Acc_p7', 'Acc_p8']].groupby('Size').mean()
```

```python
Data = pd.DataFrame(columns=['Test_on','data_size','Accuracy'])
for i in range(len(Partcp_performance[Partcp_performance['Model']=="Group"])):
    data = pd.DataFrame(columns=['Test_on','data_size','Accuracy'])
    data['Test_on'] = Partcp_performance.columns[3:]
    data['Accuracy'] = Partcp_performance[Partcp_performance['Model']=="Group"].transpose().values[3:][:,i]
    data['data_size'] = Partcp_performance[Partcp_performance['Model']=="Group"].transpose().values[1][i]
    Data = Data.append(data)
Data = Data.reset_index(drop =True)
Data.Accuracy = Data.Accuracy.astype(float)
plt.figure(figsize=(15,4))
sns.lineplot(data=Data, x = "Test_on", y = "Accuracy",hue = "data_size",marker='o')
plt.xticks(range(6),["1-5","1","2","3","4","5"],rotation = 'horizontal')
plt.ylabel("Accuracy",fontsize=12)
plt.xlabel("Speaker",fontsize=12)
plt.ylim([0.8,1.01])
plt.title("Pariticipant performance of group model")
plt.box(False)
plt.show()
```

```python
Data = pd.DataFrame(columns=['Test_on','data_size','Accuracy'])
for i in range(len(Command_performance[Command_performance['Model']=="Group"])):
    data = pd.DataFrame(columns=['Test_on','data_size','Accuracy'])
    data['Test_on'] = Command_performance.columns[3:]
    data['Accuracy'] = Command_performance[Command_performance['Model']=="Group"].transpose().values[3:][:,i]
    data['data_size'] = Command_performance[Command_performance['Model']=="Group"].transpose().values[1][i]
    Data = Data.append(data)
Data = Data.reset_index(drop =True)
Data.Accuracy = Data.Accuracy.astype(float)
plt.figure(figsize=(15,4))
sns.lineplot(data=Data, x = "Test_on", y = "Accuracy",hue = "data_size",marker='o')
plt.xticks(range(9),["1-5","1","2","3","4","5","6","7","8"],rotation = 'horizontal')
plt.ylabel("Accuracy",fontsize=12)
plt.xlabel("Speaker",fontsize=12)
# plt.ylim([0.8,1.01])
plt.title("Command performance of group model")
plt.box(False)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(Command_performance[['Size','Command_Acc_p15','Command_Acc_p1', 'Command_Acc_p2',
                                  'Command_Acc_p3', 'Command_Acc_p4', 'Command_Acc_p5', 'Acc_p6', 'Acc_p7', 'Acc_p8']].groupby('Size').mean(),
            annot = True,cmap='Greens')
plt.yticks(ticks =  [0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5],labels = Command_performance[['Size','Command_Acc_p15','Command_Acc_p1', 'Command_Acc_p2',
                                  'Command_Acc_p3', 'Command_Acc_p4', 'Command_Acc_p5', 'Acc_p6', 'Acc_p7', 'Acc_p8']].groupby('Size').mean().index, rotation = 'horizontal')
plt.xticks(ticks = [0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5],labels = ["P1-5","P1","P2","P3","P4","P5","P6","P7","P8"],rotation = 'horizontal')
plt.ylabel("Training Data Size")
plt.xlabel("Accuracy on Command")
plt.title("Command Accuray")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(Partcp_performance[['Size','Partcp_Acc_p15','Partcp_Acc_p1', 'Partcp_Acc_p2',
                                  'Partcp_Acc_p3', 'Partcp_Acc_p4', 'Partcp_Acc_p5']].groupby('Size').mean(),
            annot = True,cmap='Greens')
plt.yticks(ticks =  [0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5],labels = Partcp_performance[['Size','Partcp_Acc_p15','Partcp_Acc_p1', 'Partcp_Acc_p2',
                                  'Partcp_Acc_p3', 'Partcp_Acc_p4', 'Partcp_Acc_p5']].groupby('Size').mean().index, rotation = 'horizontal')
plt.xticks(ticks = [0.5,1.5,2.5,3.5,4.5,5.5],labels = ["P1-5","P1","P2","P3","P4","P5"],rotation = 'horizontal')
plt.ylabel("Training Data Size")
plt.xlabel("Accuracy on Command")
plt.title("Speaker Accuray")
plt.show()
```
